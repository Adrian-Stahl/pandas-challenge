```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Total number of unique schools
schoolnames = school_data_complete['school_name'].unique()
total_schools = len(schoolnames)
total_schools 
```

```python
#Total students
total_students = school_data_complete['student_name'].count()
total_students
```

```python
#Total budget
total_budget = school_data['budget'].sum()
total_budget
```

```python
#Average math score
avg_math = school_data_complete['math_score'].mean()
avg_math
```

```python
#Average reading score
avg_read = school_data_complete['reading_score'].mean()
avg_read
```

```python
#% passing math (the percentage of students who passed math)
students_passmath = school_data_complete.loc[school_data_complete['math_score']>=70]
numstudents_passmath = students_passmath['Student ID'].count()
percent_passmath = (numstudents_passmath/total_students)*100
percent_passmath
```

```python
#% passing reading (the percentage of students who passed reading)
students_passreading = school_data_complete.loc[school_data_complete['reading_score']>=70]
numstudents_passreading = students_passreading['Student ID'].count()
percent_passreading = (numstudents_passreading/total_students)*100
percent_passreading
```

```python
#% overall passing (the percentage of students who passed math AND reading)
pass_both = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)]['Student ID'].count()/total_students*100
pass_both
```

```python
#Create a new DataFrame for the above calculations called district_summary

district_summary = pd.DataFrame({
    "Total Number of Schools": total_schools, 
    "Total Number of Students": total_students,
    "Total Budget": total_budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "Percent Passing Math": percent_passmath,
    "Percent Passing Reading": percent_passreading,
    "Percent Passing Overall": pass_both
}, index=[0])

district_summary
```

```python
#Perform the necessary calculations and then create a DataFrame that summarizes key metrics about each school.
#Include the following:
#School name
school_name= school_data_complete.set_index('school_name').groupby(['school_name'])
```

```python
#School type
schooltype = school_data.set_index('school_name')['type']
schooltype
```

```python
#Total students

school_totalstudents = school_name['Student ID'].count()
school_totalstudents
```

```python
#Total school budget

school_total_budget = school_data.set_index('school_name')['budget']
school_total_budget
```

```python
#Per student budget
school_perstudent_budget = (school_data.set_index('school_name')['budget']/ school_data.set_index('school_name')['size'])
school_perstudent_budget
```

```python
#Average math score
school_avgmath = school_name["math_score"].mean()
school_avgmath
```

```python
#Average reading score
school_avgreading = school_name["reading_score"].mean()
school_avgreading
```

```python
#% passing math (the percentage of students who passed math)
school_percent_passmath = school_data_complete[school_data_complete['math_score']>= 70].groupby('school_name')['Student ID'].count()/school_totalstudents*100
school_percent_passmath
```

```python
#% passing reading (the percentage of students who passed reading)
school_percent_passreading = school_data_complete[school_data_complete['reading_score']>= 70].groupby('school_name')['Student ID'].count()/school_totalstudents*100
school_percent_passreading
```

```python
#% overall passing (the percentage of students who passed math AND reading)
school_passboth = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)].groupby('school_name')['Student ID'].count()/school_totalstudents*100
school_passboth
```

```python
#Create a new DataFrame for the above calculations called per_school_summary 

per_school_summary = pd.DataFrame({
    "School Name": school_name,
    "School Type": schooltype,
    "Total Students": school_totalstudents,
    "Per Student Budget": school_perstudent_budget,
    "Total School Budget": school_total_budget,
    "Average Math Score": school_avgmath,
    "Average Reading Score": school_avgreading,
    '% Passing Math': school_percent_passmath,
    '% Passing Reading': school_percent_passreading,
    "% Overall Passing": school_passboth
})
per_school_summary = per_school_summary[["School Type",
    "Total Students",
    "Per Student Budget",
    "Total School Budget",
    "Average Math Score",
    "Average Reading Score",
    "% Passing Math",
    "% Passing Reading",
    "% Overall Passing"
]]
per_school_summary.style.format ({'Total Students':'{:}',
    'Per Student Budget':'${:.2f}',
    'Total School Budget':'${:,.2f}',
    "Average Math Score":"{:2f}",
    "Average Reading Score":"{:2f}",
    "% Passing Math":"{:2f}",
    "% Passing Reading":"{:2f}"
})
```

```python
#Sort the schools by % Overall Passing in descending order and display the top 5 rows.
#Save the results in a DataFrame called "top_schools".


top_schools = per_school_summary.sort_values("% Overall Passing", ascending = False)
top_schools.head().style.format({'Total Students':'{:}',
    'Per Student Budget':'${:.2f}',
    'Total School Budget':'${:,.2f}',
    "% Passing Math":"{:2f}",
    "% Passing Reading":"{:2f}",
    "% Overall Passing":"{:2f}"
})
```

```python
# Lowest-Performing Schools (by % Overall Passing)
# Sort the schools by % Overall Passing in ascending order and display the top 5 rows.
# Save the results in a DataFrame called "bottom_schools".


bottom_schools = top_schools.tail()
bottom_schools = bottom_schools.sort_values("% Overall Passing")
bottom_schools.style.format ({'Total Students':'{:}',
    'Per Student Budget':'${:.2f}',
    'Total School Budget':'${:,.2f}',
    "Average Math Score":"{:2f}",
    "Average Reading Score":"{:2f}",
    "% Passing Math":"{:2f}",
    "% Passing Reading":"{:2f}"
})
```

```python
#Math Scores by Grade
#Perform the necessary calculations to create a DataFrame that lists the average math score for students of each grade level (9th, 10th, 11th, 12th) at each school.
# Use the code provided to separate the data by grade (1 points)
# Group by "school_name" and take the mean of each (4 points)
# Use the code to select only the math_score (1 points)
# Combine each of the scores above into single DataFrame called math_scores_by_grade (4 points)
math_freshmen = student_data.loc[student_data['grade'] == '9th'].groupby('school_name')['math_score'].mean()
math_sophomore = student_data.loc[student_data['grade'] == '10th'].groupby('school_name')['math_score'].mean()
math_junior = student_data.loc[student_data['grade'] == '11th'].groupby('school_name')['math_score'].mean()
math_senior = student_data.loc[student_data['grade'] == '12th'].groupby('school_name')['math_score'].mean()

math_scores_by_grade = pd.DataFrame({'9th':math_freshmen, 
                            '10th':math_sophomore,
                            '11th':math_junior,
                            '12th':math_senior })

math_scores_by_grade = math_scores_by_grade[['9th','10th','11th','12th']]
math_scores_by_grade.index.name = "School Name"
math_scores_by_grade.style.format({'9th':'{:.2f}',
                           '10th':'{:.2f}',
                           '11th':'{:.2f}',
                           '12th':'{:.2f}'})
```

```python
# Reading Scores by Grade (10 points)
# Use the code provided to separate the data by grade (1 points)
# Group by "school_name" and take the mean of each (4 points)
# Use the code to select only the reading_score (1 points)
# Combine each of the scores above into single DataFrame called reading_scores_by_grade (4 points)
reading_freshmen = student_data.loc[student_data['grade'] == '9th'].groupby('school_name')['reading_score'].mean()
reading_sophomore = student_data.loc[student_data['grade'] == '10th'].groupby('school_name')['reading_score'].mean()
reading_junior = student_data.loc[student_data['grade'] == '11th'].groupby('school_name')['reading_score'].mean()
reading_senior = student_data.loc[student_data['grade'] == '12th'].groupby('school_name')['reading_score'].mean()

reading_scores_by_grade = pd.DataFrame({'9th':reading_freshmen, 
                            '10th':reading_sophomore,
                            '11th':reading_junior,
                            '12th':reading_senior })
reading_scores_by_grade = reading_scores_by_grade[['9th','10th','11th','12th']]
reading_scores_by_grade.index.name = "School Name"

reading_scores_by_grade.style.format({'9th':'{:.2f}',
                            '10th':'{:.2f}',
                            '11th':'{:.2f}',
                            '12th':'{:.2f}'})
```

```python
# Scores by School Spending
# Create a table that breaks down school performance based on average spending ranges (per student).
# Use the scores above to create a DataFrame called spending_summary.
# Include the following metrics in the table:
# Average math score
# Average reading score
# % passing math (the percentage of students who passed math)
# % passing reading (the percentage of students who passed reading)
# % overall passing (the percentage of students who passed math AND reading
# Use the code provided below to create four bins with reasonable cutoff values to group school spending.



spending_bins = [0, 585, 630, 645, 680]
category = ["<$585", "$585-630", "$630-645", "$645-680"]

school_data_complete["Spending Ranges (Per Student)"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], spending_bins, labels = category) 

grp_byspending = school_data_complete.groupby("Spending Ranges (Per Student)")

spending_math_scores = grp_byspending['math_score'].mean()
spending_reading_scores = grp_byspending['reading_score'].mean()
spending_passing_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby("Spending Ranges (Per Student)")['Student ID'].count()/grp_byspending['Student ID'].count()*100
spending_passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70].groupby("Spending Ranges (Per Student)")['Student ID'].count()/grp_byspending['Student ID'].count()*100
overall_passing_spending = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby("Spending Ranges (Per Student)")['Student ID'].count()/grp_byspending['Student ID'].count()*100

spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending 
})
spending_summary.index.name = "Per Student Budget"
spending_summary.style.format({"Average Math Score":'{:.2f}',
    "Average Reading Score":'{:.2f}',
    "% Passing Math":'{:.2f}',
    "% Passing Reading": '{:.2f}',
     "% Overall Passing":'{:.2f}'                                      
})

```

```python
# Scores by School Size

size_bins = [0, 1000, 2000, 5000]
category = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_data_complete["bins_size"] = pd.cut(school_data_complete['size'], size_bins, labels = category
                                           
bin_bysize = school_data_complete.groupby("bins_size")
averagemath_score = bin_bysize["math_score"].mean()
averagereading_score = bin_bysize["reading_score"].mean() 
math_percent_pass = school_data_complete[school_data_complete['math_score'] >= 70].groupby('bins_size')['Student ID'].count()/bin_bysize['Student ID'].count()*100                                           
reading_percent_pass =  school_data_complete[school_data_complete['reading_score'] >= 70].groupby('bins_size')['Student ID'].count()/bin_bysize['Student ID'].count()*100  
overall_pass = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('bins_size')['Student ID'].count()/bin_bysize['Student ID'].count()*100                                      

size_summary = pd.DataFrame({
    "Average Math Score": averagemath_score,
    "Average Reading Score": averagereading_score,
    "% Passing Math": math_percent_pass,
    "% Passing Reading": reading_percent_pass,
    "% Overall Passing": overall_pass})       
size_summary.index.name = "Total"  

spending_summary.style.format({"Average Math Score":'{:.2f}',
    "Average Reading Score":'{:.2f}',
    "% Passing Math":'{:.2f}',
    "% Passing Reading": '{:.2f}',
     "% Overall Passing":'{:.2f}'                                      
})                                         
                                           
```

```python
# Scores by School Type
# Use the per_school_summary DataFrame from the previous step to create a new DataFrame called type_summary.
# This new DataFrame should show school performance based on the "School Type".


school_type = school_data_complete.groupby("type")
averagemath_score = school_type["math_score"].mean()
averagereading_score = school_type["reading_score"].mean() 
math_percent_pass = school_data_complete[school_data_complete['math_score'] >= 70].groupby('type')['Student ID'].count()/school_type['Student ID'].count()*100
reading_percent_pass = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('type')['Student ID'].count()/school_type['Student ID'].count()*100
overall_pass = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('type')['Student ID'].count()/school_type['Student ID'].count()*100

type_summary = pd.DataFrame({"Average Math Score": averagemath_score,
    "Average Reading Score": averagereading_score,
    "% Passing Math": math_percent_pass,
    "% Passing Reading": reading_percent_pass,
    "% Overall Passing": overall_pass}) 

type_summary.index.name = "Type of School"

type_summary.style.format({"Average Math Score":'{:.2f}',
    "Average Reading Score":'{:.2f}',
    "% Passing Math":'{:.2f}',
    "% Passing Reading": '{:.2f}',
     "% Overall Passing":'{:.2f}'                                      
})                                
```
